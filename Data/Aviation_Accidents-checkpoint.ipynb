## Data Preparation
Nettoyage des données manquantes :

Suppression ou imputation des valeurs absentes selon le contexte

Priorité donnée aux colonnes critiques pour l’analyse (modèle, sévérité, date)

Conversion des types de données :

Transformation des colonnes de dates au bon format (datetime)

Uniformisation des champs numériques (nombre de blessés, morts, etc.)

Encodage des variables catégorielles si nécessaire

Par modèle d’avion pour comparer la fréquence et la gravité des incidents

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AviationData.csv', encoding='cp1252', low_memory=False)

# Affichage des premières lignes
display(df.head(5))
```

 ## Analyse des données manquantes

```python
# Analyse des données manquantes
missing_data = pd.DataFrame({
    'Valeurs_manquantes': df.isna().sum(),
    'Pourcentage': (df.isna().mean() * 100).round(1)
})
print("Conseil : Les colonnes avec >30% de valeurs manquantes peuvent nécessiter un traitement spécial")
display(missing_data.sort_values('Valeurs_manquantes', ascending=False).head(10))
```

## Nettoyage des données manquantes

```python
# Nettoyage des données
df_clean = df.dropna(subset=['Event.Date', 'Location']).copy()

# Conversion et traitement des dates
df_clean['Event.Date'] = pd.to_datetime(df_clean['Event.Date'], errors='coerce')

# Gestion des valeurs manquantes pour la phase de vol
df_clean['Broad.phase.of.flight'] = df_clean['Broad.phase.of.flight'].fillna('UNKNOWN')

print("Nettoyage terminé")
print(f"Lignes restantes après nettoyage : {len(df_clean):,}")
print(f"Colonnes conservées : {len(df_clean.columns)}")
```

## Priorité donnée aux colonnes critiques pour l’analyse (modèle, sévérité, date)

```python
# Liste de colonnes prioritaires
priority_cols = ['Model', 'Injury.Severity', 'Event.Date']

# Ajouter les autres colonnes (sans doublons)
other_cols = [col for col in df_clean.columns if col not in priority_cols]

# Réorganiser le DataFrame
df_clean = df_clean[priority_cols + other_cols]
```

```python
# Détection et suppression des doublons
dup_count = df.duplicated().sum()
df = df.drop_duplicates()

print(f"Nombre de doublons supprimés : {dup_count}")
print(f"Nouvelle taille du dataset : {len(df):,} lignes")
```

## Regroupement par type météo avec moyenne de la gravité

```python
grouped_weather = df_clean.groupby('Weather.Condition')['Severity_Score'].mean().sort_values()
print(grouped_weather)
```

```python
df_clean['Severity_Score'] = (
    df_clean['Total.Fatal.Injuries'].fillna(0) * 3 +
    df_clean['Total.Serious.Injuries'].fillna(0) * 2 +
    df_clean['Total.Minor.Injuries'].fillna(0) * 1
)
```

# Conversion de la colonne de dates

```python
# Conversion de la colonne de dates
df_clean['Event.Date'] = pd.to_datetime(df_clean['Event.Date'], errors='coerce')

# Vérification des résultats
date_conversion_report = pd.DataFrame({
    'Exemple de dates': [
        df_clean['Event.Date'].min(),
        df_clean['Event.Date'].median(),
        df_clean['Event.Date'].max()
    ],
    'Description': ['Date la plus ancienne', 'Date médiane', 'Date la plus récente']
}).set_index('Description')

print("Conversion terminée. Statistiques temporelles :")
display(date_conversion_report)

# Analyse des échecs de conversion
failed_conversions = df_clean['Event.Date'].isna().sum() - df['Event.Date'].isna().sum()
if failed_conversions > 0:
    print(f"{failed_conversions} dates n'ont pas pu être converties")
```

# Uniformisation des champs numériques (nombre de blessés, morts, etc.)


```python
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries',
               'Total.Minor.Injuries', 'Total.Uninjured']

df_clean[injury_cols] = df_clean[injury_cols].fillna(0)
df_clean[injury_cols] = df_clean[injury_cols].astype(int)
```

## Agrégation des données :

Par modèle d’avion

```python
model_agg = df_clean.groupby('Model').agg({
    'Event.Id': 'count',
    'Severity_Score': 'mean',
    'Total.Fatal.Injuries': 'sum'
}).rename(columns={'Event.Id': 'Nombre_accidents'})

model_agg = model_agg.sort_values(by='Nombre_accidents', ascending=False)
```

## Calcul du nombre d'accidents par modèle d'avion

```python
# Calcul du nombre d'accidents par modèle d'avion
accidents_per_model = df['Model'].value_counts()

# Affichage des 20 premiers résultats
print(accidents_per_model.head(20))
```

## Taux d’accidents par type d’avion :


```python
# Calcul du nombre d'accidents par catégorie
accidents_by_category = df['Aircraft.Category'].value_counts().sort_values(ascending=False)

print("🔹 Nombre d'accidents par type d’avion :")
print(accidents_by_category)
```

## Identification des modèles présentant le moins d’accidents ou de blessures graves

```python
# Nombre d'accidents par modèle
accidents_per_model = df['Model'].value_counts()

# Afficher les 10 modèles avec le moins d’accidents (mais qui apparaissent au moins 5 fois)
least_accidents = accidents_per_model[accidents_per_model >= 5].sort_values().head(10)

print("🔹 Modèles avec le moins d'accidents (au moins 5 occurrences) :")
print(least_accidents)
```

```python
# Compter les accidents par modèle
accidents_per_model = df['Model'].value_counts()

# Filtrer les modèles ayant au moins 2 accidents
models_low_accidents = accidents_per_model[accidents_per_model >= 2].sort_values().head(10)

print("Modèles avec le moins d’accidents (au moins 2 incidents) :")
print(models_low_accidents)
```

```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
```

```python
# Extraire le mois
df['Month'] = df['Event.Date'].dt.month

# Visualisation
plt.figure(figsize=(7,3))
df['Month'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Nombre d'accidents par mois")
plt.xticks(range(12), ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Juin', 'Juil', 'Août', 'Sep', 'Oct', 'Nov', 'Déc'])
plt.show()
```

 Taux d'accidents par million de vols

```python
taux_mortalite = (df['Total.Fatal.Injuries'] > 0).mean() * 100
print(f" {taux_mortalite:.1f}% des accidents ont causé au moins un décès")

# Visualisation
plt.figure(figsize=(6,6))
plt.pie([taux_mortalite, 100-taux_mortalite], 
        labels=['Accidents mortels', 'Autres'], 
        colors=['#e74c3c', '#3498db'], 
        autopct='%1.1f%%')
plt.title("Proportion d'accidents mortels")
plt.show()
```

```python
# Comptage des accidents par condition météo
accidents_par_meteo = df['Weather.Condition'].value_counts()

# Les 10 conditions météo les plus fréquentes
top_conditions = accidents_par_meteo.head(10)

plt.figure(figsize=(12,6))
top_conditions.plot(kind='bar', color='cornflowerblue')
plt.title("Nombre d'accidents par condition météorologique")
plt.xlabel("Condition météo")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
